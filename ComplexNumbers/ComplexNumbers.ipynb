## Комплексно число

### Дефиниция
**Комплексно число** е всяко число във следната форма:
$$
    z = a+ib
$$
Където $a$ и $b$ са **реални числа** и $i$ е **имагинерната единица**.

Числото $a$ се нарича **реална част** $z$, а числото $b$ се нарича **имагинерна част** на $z$.
Означават се съответно $a = Re(z)$ и $b = Im(z)$

Това представяне $a+ib$ на комплексните числа се нарича **алгебричен вид** на комплексно число.

Често числата се записват и така 
$$
    z = a+bi
$$

Примери
$$
4-9i, \,\, 2+3i, \,\, -5+0i = -5, \,\,0 + 8i = 8i, \,\,-9 + 4i
$$

Числата чиито реална част е нула се наричан **чисто имагинерни**, такова е например $8i$.

Реалните числа естествено се съдържат в множеството на комплексните, всяко реално число $a$ се представя като комплексно със имагинерна част 0 и реална част самото число.

$$
    a = a + 0i
$$

С **R** означаваме множеството на реалните числа а с **C** множеството на комплексните, от последното се вижда че множеството на реалните се включва в множеството на комплексните.

### Равенство на комплексни числа
Имаме равенство на две комплексни числа когато съответните им реални и имагинерни части съвпадат.

#### Дефиниция
Комплексните числа $z_1 = a_1 + ib_1$ и $z_2 = a_2 + ib_2$ са равни, $z_1 = z_2$, ако $a_1 = a_2$ и $b_1 = b_2$.

Всяко комплексно число има единствено представяне чрез реалната си и имагинерната си част. Две комплескни числа са равни ако реалните им и комплексните им части са равни.

 
### Операции със комплексни числа
Да си изведем правилата за събиране, изваждане, умножение и деление на комплексни числа

Нека
$z_1 = a_1 + ib_1$ и $z_2 = a_2 + ib_2$

#### Събиране

Събирането става като събираме съответните реални и имагинерни части - (събираме еднакви с еднакви).

$$
  z_1 + z_2 = (a_1+a_2) + i(b_1+b_2)
$$


#### Умножение

Умножението правим все едно умножаваме 2 полинома в които участва $i$, и използваме че $i^2=-1$

 $$
  z_1 z_2 \, = \, (a_1 + ib_1)(a_2 + ib_2) \, = \, a_1a_2 + i b_1 a_2 + a_1 i b_2 + i b_1 i b_2 \, = \, 
  a_1a_2 + i(b_1 a_2 + a_1 b_2) + i^2 b_1 b_2 \, = \, 
  a_1a_2 + i(b_1 a_2 + a_1 b_2) - b_1 b_2 
 $$
 
така получаваме
 $$
  z_1 z_2 = (a_1a_2 - b_1 b_2) + i(b_1 a_2 + a_1 b_2)
 $$
 
 Дотук сме дефинирали 2 операции със комплексни числа събиране и умножение


#### Комутативност, Асоциативност Дистрибутивност

За домашна проверете следните свойства:

Комутативност на събирането и умножението

$$
    z_1+z_2 = z_2+z_1
$$

$$
    z_1 z_2 = z_2 z_1
$$

Асоциативност на събирането и умножението

$$
    z_1+(z_2+z_3) = (z_1+z_2)+z_3
$$

$$
    z_1(z_2 z_3) = (z_1z_2)z_3
$$

Дистрибутивност на събирането и умножението

$$
    z_1(z_2+z_3) = z_1z_2+z_1z_3
$$

#### Нулев и Единичен Елемент/Число

Числото $0 = 0 + i0$ играе ролята на 0 за комплесните числа и съвпада с 0 на реалните.

$$
    0 + z = z + 0 = a + ib + 0 + i0 = (a+0) + i(b+0) = a+ib = z
$$

Числото $1 = 1 + i0$ играе ролята на 1 за комплесните числа и съвпада с 1 на реалните.

$$
    1.z = z.1 = (a + ib)(1 + i0) = (a.1 - b.0) + i(b.1 + a.0) = a + ib = z 
$$


#### Изваждане

За дадено комплексно $z = a+ib$ число, числото 
    $$-z = (-a) + i(-b) = -a - ib$$ ще наричаме **противоположно число** или обратно по отношение на събирането число. Тъй като
    
$$
    z + (-z) = (a+(-a)) + i(b+(-b)) = 0+i0 = 0
$$

Изваждането ще го дефинираме като събиране със противоположното на умалителя:

$$
    z_1 - z_2 = z_1 + (-z_2) = a_1+ib_1 + (-a_2) + i(-b_2) = (a_1-a_2) + i(b_1-b_2) 
$$

На практика изваждането става като извадим съответните реални и имагинерни часни.

$$
  z_1 - z_2 = (a_1-a_2) + i(b_1-b_2)
$$


#### Числов Пръстен, За да стане Числово Поле ни трябва и операция деление.
Дотук така дефинираните операции и техните свойства определят множеството от комплескните като Числов Пръстен, за да стане Числово поле ни трябва още една операция - делене. Или по-специално трябва да дефинираме за дадено число $z \neq 0$ неговото **реципрочно число** $z^{-1}$ така че $z^{-1}z = zz^{-1} = 1$.

Преди да дефинираме как ще търсим реципрочното ще дефинираме какво е **комплексно спрегнато**. 

#### Дефиниция - Комплексно Спрегнато
Ако имаме комплексното число $$ z = a + ib $$ то числото $$ \overline{z} = a - ib $$ ще наричаме **комплексно спрегнато**. 

Комплексно спрегнатото на дадено комплексно число се получава като сменим знака на имагинерната част на числото.

Пример
$$
    z = 6+3i $$ $$
    \overline z = 6-3i
$$
Пример
$$
    z = -5+i $$ $$
    \overline z = -5-i
$$

Комплексно спрегнатото на реално число е същото реално число

$$
    a = a+0i = a $$ $$
    \overline a = a -0i = a
$$

Комплексно спрегнатото на комплексно спрегнатото дава първоначалното число

$$
    \overline{\overline{z}} = z
$$
Тъй като 
$$
    z = a+ib $$ $$
    \overline{z} = a+(-b)i $$ $$
    \overline{\overline{z}} = a+(-(-b))i = a+ib = z
$$

Комплексно спрегнатото на сборът, разликата и произведението на комплексно спрегнати е сборът, разликата и произведението от комплексно спрегнаните.

$$
    \overline{z_1 + z_2} = \overline{z_1} + \overline{z_2} $$ $$
    \overline{z_1 - z_2} = \overline{z_1} - \overline{z_2} $$ $$
    \overline{z_1 \, z_2} = \overline{z_1} \, \overline{z_2}
$$

Доказателство за произведението, за събирането и изваждането са тривиални.

$$
    z_1 = a + ib, \,\, 
    z_2 = c + id $$ $$
    \overline{z_1} = a - ib, \,\,  
    \overline{z_2} = c - id 
$$
$$
\overline{z_1.z_2} = \overline{(ac-bd) + i(ad+bc)} = (ac-bd) - i(ad+bc) $$ $$
\overline{z_1} . \overline{z_2} = (ac-(-b)(-d)) + i(a(-d)+(-b)c) = (ac-bd) - i(ad+bc) = \overline{z_1.z_2}
$$

Сборът на едно число с неговото комплексно спрегнато дава 2 пъти реалната част на числото - реално число, а разликата на едно число със комплексно спрегнатото дава чисто имагинерно число равно на 2 пъти имагинерната част по имагинерната единица

$$
    z = a + ib $$ $$
    \overline{z} = a - ib 
$$

$$
    z + \overline{z} = ( a + ib ) + ( a - ib ) = 2a  
$$ 
$$
    z - \overline{z} = ( a + ib ) - ( a - ib ) = ( a - a ) + i( b -(-b) ) = i2b
$$

В такъв случай можем да кажем че

$$
    Re(z) = \frac{ z + \overline{z} }{2} 
$$ 
$$
    Im(z) = \frac{ z - \overline{z} }{2i}
$$

Произведението на едно число по неговото комплексно спрегнато е реално неотрицателно и има много важен смисъл. Когато разгледаме геометричното представяне на комплексните числа това ще се окаже квадрата на дължината на вектора с начало координатната система и край комплексното число.

$$
    z = a + ib  $$ $$
    \overline{z} = a - ib
$$ 

$$
    z\overline{z} = (a + ib)(a - ib) = a^2-(ib)^2 = a^2-i^2b^2 = a^2-(-1)b^2 = a^2 + b^2
$$ 

или можехме просто да заместим във правилото за умножение

$$
  z\overline{z} = (a + ib)(a - ib) = (aa - b(-b)) + i(ba + a(-b)) = a^2+b^2 + i(ab-ab) = a^2+b^2 + i0 = a^2+b^2
$$

Така важният резултат от произведението на едно число по неговото комплексно спрегнато е

$$
  z\overline{z} = a^2+b^2
$$

#### Дефиниция Абсолютна Стойност на Комплексно Число 
**Абсолютната стойност** или още **модула** на едно комплексно число се задава така 
$$
  \left| z \right| = \sqrt{z\overline{z}}
$$

Вижда се че е корен квадратен от сборът от квадратите на реалната и имагинерната част, точно както в Питагоровата теорема 
$$
  \left| z \right| = \sqrt{z\overline{z}} = \sqrt{a^2+b^2}
$$

Модулите на комплексно и неговото комплексно спрегнато са равни
$$
  \left| \overline{z} \right| = \sqrt{\overline{z}\,\overline{\overline{z}}} = \sqrt{\overline{z}\,z} = \sqrt{z\overline{z}} = \left| z \right| = \sqrt{a^2+b^2}
$$


#### Деление на Комплексни Числа

Да отбележим че, можем да делим комплексно на реално число като го умножим по реципрочното на реалното число. Нека $a$, $b$ и $c$ са реални числа

$$
    \frac{(a+ib)}{c} = \frac{1}{c}(a+ib) = \frac{a}{c}+i\frac{b}{c}
$$

Готови сме да дефинираме делението на комплексни числа.

$$
\frac{z_1}{z_2}
$$

Трябва да се освободим от комплексното число в знаменателя, така че за знаменател да получим реално число. Идеята е да умножим числителя и знаменателя по комплексно спрегнатото на знаменателя, така знаменателя ще се превърне в реално положително число. Ще използваме че, всяко число различно от 0 разделено на себе си трябва да дава 1, а произведението на 2 дроби е произведението на числителите разделено на произведението на знаменателите. 

$$
    \frac{z_1}{z_2} =  \frac{z_1}{z_2} \cdot \frac{\overline{z_2}}{\overline{z_2}} = \frac{z_1 \overline{z_2}}{z_2\overline{z_2}}
$$

#### Частното на Комплексни Числа
$$
\frac{z_1}{z_2} = \frac{z_1 \overline{z_2}}{z_2\overline{z_2}}
$$

#### Дефиниция на Реципрочно на Комплексно Число
$$
    z^{-1} = \frac{1}{z} = \frac{\overline{z}}{z\overline{z}} = \frac{1}{z\overline{z}}\overline{z} = \frac{1}{\left| z \right|^2}\overline{z}
$$

Доказателство че числото е реципрочно 

$$
    zz^{-1} = z\frac{1}{z\overline{z}}\overline{z} = \frac{1}{z\overline{z}}z\overline{z} = 1
$$

#### Деленето формално въведено като умножение със реципрочното
От дефиницията към правилото за деление, за теорията е важно как играждаме понятията, за практиката на смятане няма значение.

$$
    \frac{z_1}{z_2} = {z_1}{z_2}^{-1} = {z_1}\frac{1}{z_2\overline{z_2}}\overline{z_2} = \frac{1}{z_2\overline{z_2}}{z_1}\overline{z_2} = \frac{{z_1}\overline{z_2}}{z_2\overline{z_2}}
$$

Последно правилото за частно на комплексно спрегнати. Както за умножението, така и за делението - комплексно спрегнатото на частното на две комплексни е частното от комплексно спрегнатите

$$
\overline{\left(\frac{z_1}{z_2}\right)} = \frac{\overline{z_1}}{\overline{z_2}} 
$$

Доказателство

$$
    \overline{\left(\frac{z_1}{z_2}\right)} = \overline{\left({z_1}\frac{1}{z_2\overline{z_2}}{\overline{z_2}}\right)} = \overline{z_1} \overline{\left(\frac{1}{z_2\overline{z_2}}\right)} \overline{\overline{z_2}} = \overline{z_1} \frac{1}{z_2\overline{z_2}} \overline{\overline{z_2}} =  \overline{z_1} \frac{1}{\overline{\overline{z_2}}\overline{z_2}} \overline{\overline{z_2}} = \frac{\overline{z_1}}{\overline{z_2}} 
$$

Тук обърнете внимание на това че числото $\frac{1}{z_2\overline{z_2}}$ e реално и затова е равно на комплексно спрегнатото си.


### Разлики с Реалните Числа

Макар в много отношения комплексните и реалните числа да си приличат, като операциите с реални числа да се пренасят за комплексни, като например свойството непрекъснатост, има и забележителни разлики - като например релацията **сравнение** на две числа няма смисъл за комплексни.

Например

## Комплексната Равнина
## Геометрично Представяне на Комплексните Числа


Както виждаме, от алгебричния вид на комплексните числа, всяко комплексно число се представя по единствен начин, чрез наредена двойка от реални числа - едното за реалната част, другото за имагинерната част. Ако комплексното число е $z = x + iy$, то на него съответства наредената двойка от числа $(x,y)$. Всяко комплескно число определя, по единствен начин една наредена двойка от реални числа. Валидно е, и обратното твърдение, всяка двойка от реални числа определя единствено комплексно. Ако двойката от числа е $(x,y)$, то на нея съответства числото $z = x + iy$ - първото от числата за реална част, а второто от числата за имагинерна част. 

Ако въведем в равнината Декартова координатна система, тогава наредените двойки от числа $(x,y)$ всъщност са координати на точки в равнината, като всяка двойка числа определя точно една точка от равнината, а всяка точка има точно една двойка координати $(x,y)$. 

На всяко комплексно число $z = x + iy$ можем по единствен начин да съпоставим точка с координати $(x,y)$ - точката с координати наредената двойка от реалната част и имагинерната част на числото - като, **реалната част за абцисата** на точката и **имагинерната част за ординатата** на точката. Също така, на всяка точка от равнината, която има координати наредената двойка $(x,y)$, можем да съпоставим единствено комплексно число $z = x + iy$, съответно с реална част абцисата на точката и имагинерна част ординатата на точката.


### Комплексна Равнина

След като има пълно съответствие между комплексно число $z = x + iy$, и точка в равнината с координати наредената двойка $(x,y)$, то ще използваме думите **комплексно число** и **точка** като синоними. 

Равнината с въведената в нея Декартова координатна система, в която са изобразени комплесните числа като точки, ще наричаме **комплексна равнина**. Хоризонталната или абцисната ос се нарича **реална ос** защото всяка точка от тази ос представя едно реално число. Вертикалната или ординатната ос се нарича **имагинерна ос** защото всяка точка от тази ос представя едно чисто имагинерно число. 


#### Геометрично представяне на комплескно число като точка

Всяко комплексно число се представя като точка в равнината с абциса реалната част на комплексното число и ордината имагинерната част на комплексното число.


Eто как изглежда геометричното представяне за $z = x + iy$, с конкретен пример $z = 3+2i$, съответно двойката $(3,2)$ са координатите на точката. 

```python
z = 3+2j

def draw_complex_plane(dimentions):
    # render axis of coord system
    ax = plt.gca()
    ax.spines["bottom"].set_position("zero")
    ax.spines["left"].set_position("zero")
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.set_aspect("equal")
    ax.axis(dimentions)
    plt.xlabel("Реална ос")
    plt.ylabel("Имагинерна ос")

def draw_point(z):
    plt.plot([z.real,z.real], [0,z.imag], 'g--')
    plt.plot([0,z.real], [z.imag,z.imag], 'g--')
    plt.plot([z.real], [z.imag], 'ko')
    
def draw_annotation(z, label):
    plt.annotate(label, xy=(z.real, z.imag), xytext=(z.real + 0.2, z.imag - 0.1))
    plt.annotate('x', xy=(z.real, 0), xytext=(z.real + 0.1, +0.1))
    plt.annotate('y', xy=(0, z.imag), xytext=(+0.1, z.imag +0.1))

draw_complex_plane([0, 5, 0, 3])
draw_point(z)
draw_annotation(z,'x + iy\n(x,y)')

plt.show()
```

### Вектори

Както вече знаем от други курсове по математика, наредените двойки от числа могат да се интерпретират като компонентите на **вектор**. Така комплексното число $z=x+iy$ може да бъде разглеждано като двуизмерен радиус-вектор спрямо началото на координатната система - това е вектор с начало, точката, която е начало на координатната система и край точката, която съответства на комплексното число - точката с координати $(x,y)$.


#### Геометрично представяне на комплескно число като двуизмерен вектор

Всяко комплексно число се представя като вектор в равнината с начало координатната система и край точката, която геометрично представя комплексното число в равнината (точката с абциса реалната част и ордината имагинерната част на числото).

Eто как изглежда геометричното представяне като вектор на $z = x + iy$, с конкретен пример $z = 3+2i$


```python
z = 3+2j

draw_complex_plane([0, 5, 0, 3])
draw_point(z)
draw_annotation(z,'x + iy\n(x,y)')
plt.quiver(0, 0, z.real, z.imag, angles = "xy", scale_units = "xy", scale = 1, color="red")
plt.show()
```

### Модул На Комплексно Число

Както е известно на всеки вектор с коордитати $(x,y)$ можем да определим **дължина на вектора** от началото на координатната система, като използваме Питагоровата теорема, ако вектора има координати $v = (x,y)$, тогава дължината му, която ще означаваме с $\left|v \right|$ e:

$$ \left|v \right| = \sqrt{x^2+y^2} $$ 

По същият начин на едно комплескно число можем да съпоставим дължината на вектора който му съответства. Тази дължина ще наричаме **модул** на комплексното число или също ще я наричаме **абсолютна стойност** на комплексното число.

#### Дефиниция Модул (Абсолютна Стойност) на Комплексно число
Ако $z = x+iy$, тогава модул на това комплескно число ще наричаме:

$$ \left| z \right| = \sqrt{x^2+y^2} $$

#### Свойства на модула

Произведението на едно комплексно с неговото комплексно спрегнато, както видяхме е

$$  z \overline z = x^2+y^2 $$

Следователно 

$$ \left| z \right|^2 = x^2+y^2 = z \overline z$$

$$ \left| z \right| = \sqrt{z \overline z}$$

Така виждаме, важното изразяване на модула на едно комлексно число, чрез произведението на числото и неговото комплексно спрегнато. Използвайки го можем да докажем следните свойства:

$$
    \left| z_1 z_2 \right| = \left| z_1 \right| \left| z_2 \right| 
$$

$$
    \left| {z}^2 \right| = \left| z \right|^2 
$$

$$
    \left| \frac{z_1}{z_2} \right| = \frac{\left| z_1 \right|}{\left| z_2 \right|} 
$$

**Доказателство** за произведението

$$
    \left| z_1 z_2 \right| = 
    \sqrt{z_1 z_2 \overline{z_1 z_2}} =
    \sqrt{z_1 z_2 \overline{z_1}\overline{z_2}} =
    \sqrt{z_1 \overline{z_1}}\sqrt{z_2 \overline{z_2}} =
    \left| z_1 \right| \left| z_2 \right| 
$$

Прилагаме го за $z = z_1 = z_2$ и получаваме

$$
    \left| {z}^2 \right| = \left|{z}{z}\right| = \left|{z}\right|\left|{z}\right| = \left| z \right|^2
$$

**Доказателство** за частното

$$
    \left| \frac{z_1}{z_2} \right| = \left(\frac{z_1}{z_2}\overline{\left(\frac{z_1}{z_2}\right)}\right)^{\frac{1}{2}} = 
    \left(\frac{z_1}{z_2}\frac{\overline{z_1}}{\overline{z_2}}\right)^{\frac{1}{2}} = 
    \left(\frac{z_1\overline{z_1}}{z_2\overline{z_2}}\right)^{\frac{1}{2}} = 
    \frac{\sqrt{z_1\overline{z_1}}}{\sqrt{z_2\overline{z_2}}} =
    \frac{\left| z_1 \right|}{\left| z_2 \right|}
$$


### Други свойства на модула

**Положителна определеност**

$ \left|z\right| \geqslant 0 $, Като $ \left|z\right| = 0 $ тогава и само тогава когато $z = 0$.

**Разширява понятието модул на реално число**

Ако $a$ e реално то $\left|a+0i\right| = \left|a\right|$, понятието модул на комплексно число включва и разширява понятието модул на реално число.


**Модула на комплесно спрегнатото** $$\left|\overline{z}\right| = \left|{z}\right|$$ 

**Модула на противоположното число** $$\left|-{z}\right| = \left|{z}\right|$$ 

**Модула на произведение на реално със комплексно число** $$\left|a{z}\right| = \left|{a}\right|\left|{z}\right|$$

### Геометричния смисъл на Събирането и Изваждането

Събирането на две комплексни числа $z_1 = a_1 + ib_1$ $z_2 = a_2 + ib_2$ се прави като се съберат съответно реалната и комплексната част на числата:

$$
    (a_1 + ib_1) + (a_2 + ib_2) = (a_1 + ib_1) + i(b_1 + b_2)
$$

Което изразено чрез наредените двойки от реална и комплексна част, които им съответстват изглежда така:

$$
    (a_1, b_1) + (a_2, b_2) = (a_1 + a_2, b_1 + b_2)
$$

Което представлява начинът, по който се събират покомпонентно вектори. Следователно събирането на комплексните числа има геометричния смисъл на **събиране на вектори** по правилото на успоредника.

#### Геометричния смисъл на събиране на комплексните числа е събиране на векторите които им съответстват. 

Пример

```python
z1 = 4+2j
z2 = 2+3j
z = z1+z2

def draw_annotation_s(z, label):
    plt.annotate(label, xy=(z.real, z.imag), xytext=(z.real + 0.3, z.imag - 0.1))

draw_complex_plane([0, 7, 0, 6])
plt.plot([z1.real], [z1.imag], 'ko')
draw_annotation_s(z1,'z1')
plt.quiver(0, 0, z1.real, z1.imag, angles = "xy", scale_units = "xy", scale = 1, color="red")

plt.plot([z2.real], [z2.imag], 'ko')
draw_annotation_s(z2,'z2')
plt.quiver(0, 0, z2.real, z2.imag, angles = "xy", scale_units = "xy", scale = 1, color="red")

plt.plot([z.real,z1.real], [z.imag,z1.imag], 'g--')
plt.plot([z.real,z2.real], [z.imag,z2.imag], 'g--')

plt.plot([z.real], [z.imag], 'ko')
draw_annotation_s(z,'z1+z2\n(x1+x2,y1+y2)')
plt.quiver(0, 0, z.real, z.imag, angles = "xy", scale_units = "xy", scale = 1, color="red")

plt.show()
```

Подобен е геометричния смисъл на изваждането - изваждане на вектори

Изваждането на две комплексни числа $z_1 = a_1 + ib_1$ $z_2 = a_2 + ib_2$ се прави като се извадят съответно реалната и комплексната част на числата, което изразено чрез наредените двойки от реална и комплексна част, които им съответстват изглежда така:

$$
    (a_1, b_1) - (a_2, b_2) = (a_1 - a_2, b_1 - b_2)
$$

Което представлява начинът, по който се изваждат покомпонентно вектори.

Също разликата на двата вектора геометрично можем за получим по правилото на триъгълника, като вектора с начало краят на умалителя и край края на умаляемото, както е показано на чертежа отдолу с примера. На чертежа в примера, разликата $z = z_2-z_1$ я получаваме като насочената отсечка с начало края на $z_1$ и край края на $z_2$. След което тази насочена отсечка я пренасяме успоредно в началото на координатната система, това е така защото всички насочени отсечки които са успоредни помежду си, равни по големина и сочещи в една посока са представители на един и същи вектор. Координатите на така пренесената отсечка са точно тези показани по-горе - покомпонентното изваждане на вектори.

### Разстояние между две точки в Комплексната Равнина

**Разстоянието между две точки** $z_1 = x_1 + iy_1$ и $z_2 = x_2 + iy_2$, е дължината на вектора, който свързва тези точки. Такъв е вектора на разликата им с начало $z_1$ и край $z_2$. 

Разстоянието между $z_1$ и край $z_2$ се дава като:
$$
\left|{z_2}-{z_1}\right| = \left|{(x_2-x_1)}+i{(y_2-y_1)}\right| = \sqrt{(x_2-x_1)^2+(y_2-y_1)^2}
$$

Което си е точно геометричното разстояние между точки със координати $(x_1,y_1)$ и $(x_2,y_2)$.

При определяне на разстоянието няма значение дали ще вземем разликата $\left|{z_2}-{z_1}\right|$ или $\left|{z_1}-{z_2}\right|$ тъй като $\left|-{z}\right|=\left|z\right|$, откъдето следва $\left|{z_2}-{z_1}\right|=\left|{z_1}-{z_2}\right|$.

### Неравенства на триъгълника

Неможем да сравняваме комплекси числа по големина но можем да сравняваме техните абсолютни стойности. Например
$z_1 = 3+4i$ и $z_2 = 5-i$, абсолютните им стойности  са $\left|{z_1}\right| = \sqrt{25} = 5$ и $\left|{z_2}\right| = \sqrt{26}$, тогава $\left|{z_1}\right| < \left|{z_2}\right|$, геометричната интерпретация е, че вектора $(3,4)$ има по-малка дължина от вектора $(5,-1)$.


Събирането на вектори освен по правилото на успоредника можем да извършваме и по правилото на триъгълника, което означава че и събирането на комплексни числа също можем да правим по правилото на триъгълника. 

Да разгледаме чертежа:


```python
z1 = 4+2j
z2 = 2+3j
z = z1+z2

def draw_annotation_s(z, label, finepos_x, finepos_y):
    plt.annotate(label, xy=(z.real, z.imag), xytext=(z.real + finepos_x, z.imag + finepos_y))

draw_complex_plane([0, 7, 0, 6])
plt.plot([z1.real], [z1.imag], 'ko')
draw_annotation_s(z1,'z1',0.3, -0.1)
plt.quiver(0, 0, z1.real, z1.imag, angles = "xy", scale_units = "xy", scale = 1, color="black")

plt.plot([z2.real], [z2.imag], 'ko')
draw_annotation_s(z2,'z2',-0.6, 0.1)
plt.quiver(0, 0, z2.real, z2.imag, angles = "xy", scale_units = "xy", scale = 1, color="blue")

plt.quiver(z1.real, z1.imag, z.real-z1.real, z.imag-z1.imag, angles = "xy", scale_units = "xy", scale = 1, color="blue")
#plt.plot([z.real,z2.real], [z.imag,z2.imag], 'g--')

plt.plot([z.real], [z.imag], 'ko')
draw_annotation_s(z,'z1+z2',0.3, 0.1)
plt.quiver(0, 0, z.real, z.imag, angles = "xy", scale_units = "xy", scale = 1, color="red")

draw_annotation_s(z/2,'|z1+z2|',-0.25, 0.9)
draw_annotation_s(z2/2,'|z2|',-0.55, 0.2)
draw_annotation_s(z1/2,'|z1|',+0.45, -0.1)
draw_annotation_s((z1+z1+z2)/2,'|z2|',0.2, -0.1)

plt.show()
```

Разглеждаме триъгълника с върхове началото - $0$ и $z_1$, $z_1+z_2$.

Дължината на отсечката с начало $0$ и край $z_1+z_2$, е равна на дължината на вектора на $z_1+z_2$, както е показано на чертежа, равна на $\left|z_1+z_2\right|$.

Дължината на отсечката с начало $0$ и край $z_1$, е равна на дължината на вектора на $z_1$, както е показано на чертежа, равна на $\left|z_1\right|$.

Отсечката с начало $z_1$ и край $z_1+z_2$, е успоредно пренесената в точка $z_1$ отсечка с начало $0$ и край $z_2$, следователно е равна на нея по дължина.

Дължината ѝ на тази отсечка с начало $z_1$ и край $z_1+z_2$, е равна на дължината на вектора на $z_2$, както е показано на чертежа, равна на $\left|z_2\right|$

#### Неравенство на триъгълника от геометрията
От геометрията знаем че дължината на една страна на триъгълник е по-малка от сборът от дължините на другите две страни, приложено за триъгълника с върхове началото - $0$ и $z_1$, $z_1+z_2$, означава че дължината на страната на триъгълника която съответства на вектора на $z_1+z_2$ е по-малка от сборът на другите две страни, които имат дължини $\left|z_1\right|$ и $\left|z_2\right|$. Така получаваме следното неравенство:

$$
\left|z_1+z_2\right|  \leqslant \left|z_1\right| + \left|z_2\right|
$$

Което е известно като **неравенство на триъгълника**


### Неравенство на триъгълника
За всеки 2 комплексни числа е изпълнено следното неравенство
$$
\left|z_1+z_2\right|  \leqslant \left|z_1\right| + \left|z_2\right|
$$
#### Доказателство
Нека $
z_1 = a_1+ib_1, \,\,\,  z_2 = a_2+ib_2,
$

$$
\left|z_1+z_2\right|^2 =  \left|(a_1+a_2)+i(b_1+b_2)\right|^2 = (a_1+a_2)^2+(b_1+b_2)^2 = {a_1}^2+ 2{a_1}{a_2} + {a_2}^2 + {b_1}^2+ 2{b_1}{b_2} + {b_2}^2 = 
$$
$$ = ({a_1}^2+{b_1}^2) + 2({a_1}{a_2}+{b_1}{b_2}) + ({a_2}^2+{b_2}^2) = \left|{z_1}\right|^2 + 2({a_1}{a_2}+{b_1}{b_2}) + \left|{z_2}\right|^2
$$

Може да се докаже самостоятелно неравенството, което е вариант на неравенството на Коши-Шварц 

$$ 
\left|{a_1}{a_2}+{b_1}{b_2}\right| \leqslant \sqrt{{a_1}^2+{b_1}^2}\sqrt{{a_2}^2+{b_2}^2} = \left|{z_1}\right|\left|{z_2}\right|
$$

Доказателството на горното неравенство може да се направи с повдигане на квадрат и отделяне на точен квадрат. 

Обратно към доказателството на неравенството на триъгълника

$$
\left|z_1+z_2\right|^2 = \left|{z_1}\right|^2 + 2({a_1}{a_2}+{b_1}{b_2}) + \left|{z_2}\right|^2 \leqslant \left|{z_1}\right|^2 + 2\left|{z_1}\right|\left|{z_2}\right| + \left|{z_2}\right|^2 = (\left|{z_1}\right| + \left|{z_2}\right|)^2
$$

Откъдето следва

$$
\left|z_1+z_2\right| \leqslant \left|{z_1}\right| + \left|{z_2}\right|
$$

  \left|z_1+z_2+z_3 + \cdots + z_n\right| \leqslant \left|z_1\right| + \left|z_2\right| + \left|z_3\right| + \cdots + \left|z_n\right|
$$

### Неравенство на Коши Буняковски Шварц (Линейна Алгебра)
Дава се за допълнителна информация **не е част** от описанието на комплексните числа, може да се пропусне.

Това неравенство е валидно за всяко линейно пространство от вектори в които е въведено скаларно произведение на два вектора. Ако в линейно пространство е въведено скаларно произведение $(a,b)$, тогава може да се въведе и норма (дължина) на вектор като $\left|{a}\right| = \sqrt{(a,a)}$. Изхождайки само от аксиомите за линейно пространство и аксиомите за скаларно произведение, се доказва следното неравенство известно като **неравенство на Коши Буняковски Шварц**

$$
    \left|(a,b)\right| \leqslant \left|a\right|\left|b\right| 
$$

Като се вземе в предвид как може да се дефинира скаларното произведение за вектори които са наредени n-орки от числа, а именно векторите представени чрез компонентите си

$$
    a =\left(a_1,a_2, \cdots, b_n\right) 
$$
$$
    b =\left(b_1,b_2, \cdots, b_n\right) 
$$

Скаларно произведение
$$
    (a,b) = a_1b_1 + a_2b_2 + \cdots + a_nb_n
$$

Норма на векторите - дължина на векторите
$$
    \left|a\right| = \sqrt{(a,a)} = {a_1}^2 + {a_2}^2 + \cdots + {a_n}^2
$$
$$
    \left|b\right| = \sqrt{(b,b)} = {b_1}^2 + {b_2}^2 + \cdots + {b_n}^2
$$

Така разписано по компоненти неравенството изглежда така:
$$ 
\left|{a_1}{b_1}+{a_2}{b_2} + \cdots + {a_n}{b_n}\right| \leqslant \sqrt{{a_1}^2+{a_2}^2 + \cdots + {a_n}^2}\sqrt{{b_1}^2+{b_2}^2 + \cdots + {b_n}^2}
$$


Виждаме че това е неравенството което ни трябваше за да докажем неравенството на триъгълника

### Косинус от ъгъл между два вектора в линейно пространство - въвежда се с неравенството на Коши Шварц 
Това неравенство ни дава възможност да въведем понятието ъгъл или по точно косинус от ъгъл между два вектора в линейно пространство ето така

$$
    cos{\theta} = \frac{(a,b)}{\left|a\right|\left|b\right|}
$$

### Следствие от Неравенство на триъгълника

От неравенството на триъгълника 

$$
\left|z_1+z_2\right|  \leqslant \left|z_1\right| + \left|z_2\right|
$$

Като се приложи за $z_1 = z_1+z_2+(-z_2)$, се получава

$$
\left|z_1\right| = \left|z_1+z_2+(-z_2)\right|  \leqslant \left|z_1+z_2\right| + \left|-z_2\right|
$$

Тъй като $\left|-z_2\right| = \left|z_2\right|$, получаваме:

$$
  \left|z_1+z_2\right| \geqslant \left|z_1\right| - \left|z_2\right|
$$

Събирането е комутативно, можем да разменим местата на $z_1$ и $z_2$, получавайки:

$$
  \left|z_2+z_1\right| \geqslant \left|z_2\right| - \left|z_1\right| = -(\left|z_1\right|-\left|z_2\right|)
$$

Обединявайки двете неравенства получаваме:

$$
  \left|z_1+z_2\right| \geqslant \left| \left|z_1\right| - \left|z_2\right| \right|
$$
